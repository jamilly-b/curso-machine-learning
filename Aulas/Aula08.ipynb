Ele possui suporte em todos os sistemas operacionais e também pode exportar gráficos para diversos formatos: pdf, svg, jpeg, png, gmp, gif, etc.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple','Google','Amazon','Microsoft','Tencent',"Mcdonald's",'Visa','Facebook','Alibaba','Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646, 611.560, 214.023, 196.526, 191.032, 186.421, 169.966, 124.273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#Redimensionando

plt.rcParams['figure.figsize'] = (15, 8) #definição geral para todos os gráficos que virão

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#conhecendo o fontsize

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize = 25) #fontsize = muda o tamanho da fonte
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
```

```python
plt.plot(empresas, bilhoes, color = 'purple') #Mudando a cor do plot
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = 15) #Adicionando uma legenda ao gráfico na posição direita
```

```python
plt.plot(empresas, bilhoes, color = 'purple', lw = 5) #aumentando a espessura da linha
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = 15)
```

```python
plt.plot(empresas, bilhoes, color = 'purple', marker = 'o')  #adicionando marcador
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = 15)
```

```python
plt.plot(empresas, bilhoes, color = 'purple', marker = 'o')  
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize = 15)
plt.grid(True)  #adicionando linhas de grade
```

```python
plt.bar(empresas, bilhoes, color = 'purple')  #mudando o tipo do grafico de barra
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'center right', fontsize = 15) #mudando a posição da legenda
plt.grid(True) 
```

```python
plt.bar(empresas,bilhoes)
```

```python
cores = ['black','dimgray','dimgrey','gray','grey','darkgray','darkgrey','silver','lightgray','lightgrey'] #criando uma lista para cores em degradê
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black') #adicionando a variável cores e colocando ua borda
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'center right', fontsize = 15) 
```

```python
plt.style.available
```

```python
plt.rcParams['figure.figsize'] = (15, 8)


plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black') 
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'center right', fontsize = 15) 
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.barh(empresas, bilhoes, color = cores, edgecolor = 'black') #barra horizontal
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Bilhões', fontsize = 20)
plt.ylabel('Empresas', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'center right', fontsize = 15) 
plt.style.use('fivethirtyeight')
plt.show()
```

```python
plt.scatter(empresas, bilhoes, color = cores, edgecolor = 'black') #grafico de dispersão
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Bilhões', fontsize = 20)
plt.ylabel('Empresas', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'center right', fontsize = 15) 
plt.style.use('fivethirtyeight')
plt.show()
```
